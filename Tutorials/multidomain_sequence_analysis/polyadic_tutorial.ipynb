```python
import pandas as pd
from sequenzo import *
import seaborn as sns
import matplotlib

# matplotlib.use("TkAgg")
import matplotlib.pyplot as plt
```

## 1. Load child and parent sequence data (assuming you have processed DataFrames from LSOG or other sources)

If you are not using the data provided by Sequenzo, instead of using load_dataset, please use  `pd.read_csv()`

```python
df_child = load_dataset("polyadic_seqc1")
df_child["id"] = range(1, len(df_child) + 1)

df_parent = load_dataset("polyadic_seqp1")
df_parent["id"] = range(1, len(df_parent) + 1)

df_child
```

## 2. Create SequenceData objects (including id, state dictionary, etc.)

```python
seq_child = SequenceData(
    data=df_child,
    time=[str(i) for i in range(15, 40)],
    states=["S", "M0", "M1", "M2", "M3+", "D"],
    labels=["S", "M0", "M1", "M2", "M3+", "D"],
    id_col="id",
    weights=None
)

seq_parent = SequenceData(
    data=df_parent,
    time=[str(i) for i in range(15, 40)],
    states=["S", "M0", "M1", "M2", "M3+", "D"],
    labels=["S", "M0", "M1", "M2", "M3+", "D"],
    id_col="id",
    weights=None
)
```

## 3. Run linked_polyad analysis

```python
result_df, merged_seqdata = linked_polyadic_sequence_analysis(seqlist=[seq_child, seq_parent],
                                                              # method="HAM",  # or "HAM", "OMspell", "CHI2"
                                                                method="OM",
                                                                distance_parameters={"sm": "CONSTANT", "indel": 1},
                                                                a=1,  # Random pairing by sequence
                                                                T=1500,
                                                                random_seed=123,
                                                                n_jobs=4,
                                                                return_df=True,
                                                                return_merged_seqdata=True,
)
```

```python
result_df
```

## What Can You Do with `result_df` *Without* a Group Comparison?

Even if your study design does **not** involve comparing groups (like gender or SES), `result_df` is still very valuable for **individual-level**, **data-driven**, or **exploratory** analysis.

### What Does Each Column in `result_df` Mean?

| Column         | Description                                                                                        |
| -------------- | -------------------------------------------------------------------------------------------------- |
| `ObservedDist` | The actual average distance between sequences in the polyad (i.e., similarity).                    |
| `U`            | The difference between random and observed distances: `U = mean_random - observed`.                |
| `V`            | The proportion of random distances greater than the observed one. A proxy for significance.        |
| `V>0.95`       | A binary indicator: 1 if `V > 0.95`, meaning the polyad is significantly more similar than random. |


### What You Can Do With It:

#### 1. **Identify highly linked polyads (significant similarity)**

Spot families with strongly shared trajectories.


```python
result_df[result_df["V>0.95"] == 1].sort_values("V", ascending=False).head()
```

**How to Interpret the Result**

The table displays polyads (family units) that show **statistically significant similarity** in life course trajectories, based on the `V > 0.95` threshold.

Each row corresponds to a polyad with the following columns:

| Column         | Interpretation                                                                                                                                       |
| -------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------- |
| `PolyadID`     | The identifier for the polyad (e.g., child + parents).                                                                                               |
| `ObservedDist` | The actual average pairwise distance between members in the polyad. Lower values indicate higher similarity.                                         |
| `U`            | Difference between the mean random distance and observed distance (`U = mean_random - observed`). Higher `U` means greater-than-expected similarity. |
| `V`            | Proportion of random distances that are greater than the observed. Values close to 1.0 indicate strong evidence of similarity.                       |
| `V>0.95`       | A binary flag indicating whether the polyads similarity is statistically significant at the 5% level.                                               |
| `Sex`          | The gender of the focal individual (e.g., child). Useful for filtering or subgroup analysis.                                                         |

Key Takeaways from This Example

* All three polyads shown have `V > 0.95`, indicating **statistically significant internal similarity**.
* Their observed distances (`4.0` or `6.0`) are low, meaning the family members' sequences are quite close.
* The `U` values are all above `21`, suggesting these observed similarities are much greater than what would be expected under random pairing.
* The presence of both "Male" and "Female" focal individuals indicates potential for follow-up comparison by gender if desired.


#### 2. **Sort or rank polyads by `U` or `V`**

Find polyads with unexpectedly low or high cohesion.

```python
result_df.sort_values("U", ascending=True).head(10)
```

Interpreting the Lowest-Ranked Polyads by **U**

These are the polyads with the **lowest U values**, meaning their observed distances are **much higher than expected by chance**. In other words, the members in these polyads have **less similar life trajectories** than expected under random pairing.

What This Table Tells Us:

* The top rows (e.g., PolyadID 25 and 37) have:

  * Very **high observed distances** (42.0),
  * Very **negative U values** (e.g., -14.6), indicating **less cohesion** than average.
  * Very **low V values** (e.g., 0.017), meaning fewer than 2% of random polyads were more dissimilar  so these are quite "dispersed."
  * `V>0.95 = 0`  **Not statistically significant similarity**.

Summary:

* These polyads likely represent families with **very different life paths** among members.
* This might reflect **weak intergenerational linkage**, or simply noise.
* Could be interesting for **contrast cases**, especially in qualitative or visual exploration.
* You can explore their sequences from `merged_seqdata` using the `PolyadID`.
